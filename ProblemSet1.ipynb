using System;
using System.Collections;

namespace triangleArea
{
    class Program
    {
        static void Main(string[] args)
        { 
           //Calling Method TriangleArea by passing 3 arguments(sides)
            TriangleArea(2, 2, 2);
            Console.ReadLine();
           
        }

        /*Declaring Method named "TriangleArea" which returns a float value,
        takes 3 paramters a,b,c of datatype float */
        static double TriangleArea(double a,
                                  double b,
                                  double c)
        
        {
            Console.WriteLine($"sides: a = {a}, b = {b}, c = {c}");

            //Calculate Semiperimeter(s)
            double s = (a + b + c) / 2;

            Console.WriteLine($"s = {s}");

            //Calculate area of triangel using Hero's formula
            double areaTriangle = (double)Math.Sqrt(s * (s - a) * (s - b) * (s - c));

            Console.WriteLine($"Area of Triangle with sides {a}, {b}, {c} is {areaTriangle}");

            return areaTriangle;
            

        }
    }
}

#### A function called frequencies() that takes a string as its only parameter, and returns a list of integers:

```python
letters = 'abcdefghijklmnopqrstuvwxyz'

# defining function frequencies() 
# which takes only a parameter of datatype string

def frequencies(message: str):
    
    letters_frequency = []
    
    # looping inside string message to fetch each charcter
    for letter in message:
        if letter in letters:  #only if a character is a letter then move ahead
            counter = message.count(letter) #counting number of times a letter repeats in a word
            letters_frequency.append(int(counter)) #adding no. of counts to a list

    print(letters_frequency) 
        
#calling the function frequencies(), 
#by passing a string "apple" 

frequencies("apple")
```
